Task 1

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Task 2

```python
uniform_distrybution = np.random.uniform(0, 1, size=(1000))
normal_distr_sigma_1 = np.random.normal(loc=0, scale=1, size=(1000))
normal_distr_sigma_10 = np.random.normal(loc=0, scale=10, size=(1000))
beta_distrybution_1 = np.random.beta(a=1, b=1, size=(1000))
beta_distrybution_2 = np.random.beta(a=2, b=2, size=(1000))
beta_distrybution_3 = np.random.beta(a=3, b=9, size=(1000))
multivariate_1 = np.random.multivariate_normal(mean=[0, 0], cov=[[1, 0], [0, 1]], size=(10000))
multivariate_2 = np.random.multivariate_normal(mean=[0, 0], cov=[[2, 1], [1, 2]], size=(10000))
```

Task 3

```python
def plot_hist(plot_size, plot_data, titles):
    fig, axes = plt.subplots(*plot_size)
    fig.set_figheight(8)
    fig.set_figwidth(15)
    axes = axes.flatten()
    for index, (plot_data, title) in enumerate(zip(plot_data, titles)):
        axes[index].hist(plot_data)
        axes[index].set_title(title)
        axes[index].grid()
    plt.show()

plot_data = [uniform_distrybution, normal_distr_sigma_1, normal_distr_sigma_10, beta_distrybution_1, beta_distrybution_2, beta_distrybution_3]
titles = ['uniform [0,1]', 'normal sigma=1', 'normal sigma=10', 'beta (a=1,b=1)', 'beta (a=2,b=2)', 'beta (a=3,b=9)']
plot_hist((2,3), plot_data, titles)
```

Task 4

```python
def plot_hist2d(plot_size, plot_data, titles, bins):
    fig, axes = plt.subplots(*plot_size)
    fig.set_figheight(15)
    fig.set_figwidth(8)
    axes = axes.flatten()
    for index, (plot_data, title, bins) in enumerate(zip(plot_data, titles, bins)):
        if bins:
            axes[index].hist2d(plot_data[:, 0], plot_data[:, 1], bins=bins)
        else:
            sns.histplot(x=plot_data[:, 0], y=plot_data[:, 1], ax=axes[index])
        axes[index].set_title(title)
        axes[index].grid()
    plt.show()

plot_data = [multivariate_1, multivariate_2] * 4
titles = ['multivariate [[1, 0], [0, 1]]', 'multivariate [[2, 1], [1, 2]]'] * 4
bins = [10, 10, 25, 25, 50, 50, None, None]
plot_hist2d((4,2), plot_data, titles, bins)
```

Task 5

```python
y_points = []
x_points = []
alphas = []
betas = []
for _ in range(1000):
    alpha = np.random.normal(2, 0.1)
    beta = np.random.normal(0.2, 0.2)
    x = np.random.normal(0, 3)
    sigma = abs(np.random.normal(0, 1))
    mikro = alpha + beta * x
    x_points.append(x)
    alphas.append(alpha)
    betas.append(beta)
    y_points.append(np.random.normal(mikro, sigma))
```

Task 6

```python
plt.scatter(x_points, y_points, c='g', s=12)
for alpha, beta in zip(alphas, betas):
    x = np.linspace(-6, 6, 1000)
    y = beta * x + alpha
    plt.plot(x, y, 'r', linewidth=0.1)
plt.xlim([-6, 6])
plt.show()
```
